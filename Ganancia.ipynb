# Problema 2

Calcular de manera numérica el cambio de la $\textbf{intensidad I(z)}$ en función de la longitud dentro del medio amplificador en los siguientes casos: 

(a) Calcular la ley de Beer no saturada.
\begin{equation}
    I(z)=I_0e^{B_0z}
\end{equation}

(b) Calcular la ley de Beer saturada.
\begin{equation}
    I(z)=I_0e^{B_0z}e^{-\frac{1}{I_{sat}}(I(z)-I_0)}
\end{equation}

con los parámetros de simulación númerica: $N=160$, $I_0=0.01$, $I_{sat}=1$, $B_0=0.5$, $\Delta z=0.1$

```python
import numpy as np
import matplotlib.pyplot as plt
#### parametros generales
N=160
Isat=1
B0=0.5
deltaz=0.1
```

```python
#### ley de beer no saturada
I0=0.01
Ins=I0
I_no_saturada=[]
for i in range(N):
    Ins=I0*np.exp(B0*i*deltaz)
    I_no_saturada.append(Ins)
```

```python
# Ley de beer saturada
I0=0.01
Is=I0
I_saturada=[]
for i in range(N):
    Is=I0*np.exp(B0*i*deltaz)*np.exp(-(Is-I0)/Isat)
    I_saturada.append(Is)
```

```python
### longitud de saturacion
zsat=deltaz*I_saturada.index(np.array(I_saturada)[np.array(I_saturada)>1][0])
zsat
```

(c) Hacer una gráfica comparativa.
(d) Ubicar en la gráfica la intensidad de saturación y la longitud de saturación.

```python
plt.plot(np.linspace(0,160*deltaz,160),I_saturada,color="blue",label="Ley de Beer saturada")
plt.plot(np.linspace(0,160*deltaz,160),I_no_saturada,color="red",label="Ley de Beer no saturada")
plt.hlines(Isat,xmin=0,xmax=16,linestyle="dashed",label="Intensidad de saturación")
plt.vlines(zsat,ymin=0,ymax=5,linestyle="dashed",color="magenta",label="Longitud de saturación")
plt.legend(loc="upper left",frameon=False)
plt.grid(True)
plt.xlabel("$\Delta z$")
plt.ylabel("$I(z)$")
plt.title("Simulacion de la ganancia en un material con $I_0=0.01$,$I_{sat}=1$,$B_0=0.5$")
plt.show()
```
